```python
import cv2
from deepface import DeepFace

# Open video capture
cap = cv2.VideoCapture(0)  # Use '0' for webcam. Change to 'path_to_video' for a video file.

# Load Haar Cascade for face detection
faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
if faceCascade.empty():
    raise ValueError("Failed to load the Haar Cascade XML file.")

while True:
    # Read a frame from the video
    ret, frame = cap.read()
    
    if not ret:
        break
    
    try:
        # Analyze the face for emotion
        result = DeepFace.analyze(frame, actions=['emotion'], enforce_detection=False)
        
        # Convert frame to grayscale for face detection
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        
        # Detect faces in the frame
        faces = faceCascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=4)
        
        for (x, y, w, h) in faces:
            # Draw rectangle around the face
            cv2.rectangle(frame, (x, y), (x+w, y+h), (255, 0, 0), 2)
            
            # Get the dominant emotion and its probability
            emotion_scores = result[0]['emotion']
            dominant_emotion = max(emotion_scores, key=emotion_scores.get)
            accuracy = emotion_scores[dominant_emotion]  # Accuracy of dominant emotion
            
            # Display the emotion and accuracy percentage on the frame
            font = cv2.FONT_HERSHEY_SIMPLEX
            display_text = f"{dominant_emotion} ({accuracy:.2f}%)"
            cv2.putText(frame, 
                        display_text, 
                        (x, y-10),  # Display text above the face rectangle
                        font, 
                        1, 
                        (0, 255, 0), 
                        2, 
                        cv2.LINE_AA)
        
        # Display the frame
        cv2.imshow('Emotion Detection', frame)
        
    except ValueError as e:
        print(f"Error: {e}")
        continue
    
    # Break the loop if 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the video capture and close windows
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
from deepface import DeepFace
from deep_sort_realtime.deepsort_tracker import DeepSort
from collections import defaultdict, Counter

# ----------------------------
# SETTINGS
# ----------------------------
BLUR = False

cap = cv2.VideoCapture(0)

faceCascade = cv2.CascadeClassifier(
    cv2.data.haarcascades + 'haarcascade_frontalface_default.xml'
)

# ----------------------------
# DEEPSORT TRACKER (CORE FIX)
# ----------------------------
tracker = DeepSort(max_age=30)  # keeps identity stable

# ----------------------------
# MEMORY
# ----------------------------
emotion_history = defaultdict(list)

# ----------------------------
# LIVE LOOP
# ----------------------------
while True:
    ret, frame = cap.read()
    if not ret:
        break

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = faceCascade.detectMultiScale(gray, 1.2, 5)

    detections = []

    # ----------------------------
    # DETECTIONS FOR DEEPSORT
    # ----------------------------
    for (x, y, w, h) in faces:

        detections.append(([x, y, w, h], 0.9, "face"))

        # optional blur
        if BLUR:
            frame[y:y+h, x:x+w] = cv2.GaussianBlur(
                frame[y:y+h, x:x+w], (35, 35), 30
            )

    # ----------------------------
    # TRACK OBJECTS
    # ----------------------------
    tracks = tracker.update_tracks(detections, frame=frame)

    for track in tracks:

        if not track.is_confirmed():
            continue

        track_id = f"Person_{track.track_id}"

        x, y, w, h = map(int, track.to_ltrb())

        face_img = frame[y:h, x:w]

        try:
            # ---------------- EMOTION ----------------
            result = DeepFace.analyze(
                face_img,
                actions=['emotion'],
                enforce_detection=False
            )

            emotions = result[0]['emotion']

            dominant = max(emotions, key=emotions.get)
            confidence = emotions[dominant]

            emotion_history[track_id].append(emotions)

            # ---------------- DISPLAY ----------------
            cv2.rectangle(frame, (x, y), (w, h), (0,255,0), 2)

            text = f"{track_id}: {dominant} ({confidence:.2f})"

            cv2.putText(frame, text, (x, y-10),
                        cv2.FONT_HERSHEY_SIMPLEX,
                        0.6, (0,255,0), 2)

        except:
            continue

    # ---------------- STATUS ----------------
    status = "BLUR ON" if BLUR else "BLUR OFF"
    cv2.putText(frame, status,
                (20, 40),
                cv2.FONT_HERSHEY_SIMPLEX,
                1,
                (0,0,255) if BLUR else (0,255,0),
                2)

    cv2.imshow("DeepSORT Emotion Tracking", frame)

    key = cv2.waitKey(1) & 0xFF

    if key == ord('q'):
        break
    elif key == ord('b'):
        BLUR = not BLUR
        print("BLUR:", BLUR)

cap.release()
cv2.destroyAllWindows()

# ----------------------------
# FINAL ANALYSIS
# ----------------------------
import matplotlib.pyplot as plt

for person_id, data in emotion_history.items():

    if len(data) == 0:
        continue

    totals = defaultdict(float)

    for frame in data:
        for emo, score in frame.items():
            totals[emo] += score

    plt.figure(figsize=(6,4))
    plt.bar(totals.keys(), totals.values())

    plt.title(f"Emotion Strength - {person_id}")
    plt.xlabel("Emotion")
    plt.ylabel("Confidence Sum")

    plt.show()
```
